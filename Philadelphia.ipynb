```python
df3 = pd.read_csv('/content/sample_data/Thomas-jefferson-university-hospital_standardcharges.csv')
```

```python
df3
```

```python
df3.info()
```

```python
cols_to_rmv = ['Modifiers','Drug_Unit_Of_Measurement','Drug_Type_Of_Measurement','Standard_Charge|Gross','Standard_Charge|Discounted_Cash','Standard_Charge|Negotiated_Percentage','Standard_Charge|Negotiated_Algorithm','Estimated_Amount','Standard_Charge|Methodology','Standard_Charge|Min','Standard_Charge|Max','Additional_Generic_Notes','Billing_Class']
df3.drop(columns=cols_to_rmv,inplace=True)
df3
```

```python
def get_cpt_code(row):
    if row['Code|1|Type'] == 'CPT':
        return row['Code|1']
    elif row['Code|2|Type'] == 'CPT':
        return row['Code|2']
    else:
        # If neither is CPT, return something else (NaN or None)
        return np.nan

df3['cpt_code'] = df3.apply(get_cpt_code, axis=1)
```

```python
df3
```

```python
cols_to_rmv = ['Code|1','Code|1|Type','Code|2','Code|2|Type']
df3.drop(columns=cols_to_rmv,inplace=True)
```

```python
df3
```

```python
df3.insert(
    loc=0,                                # position to insert at (0 = first column)
    column="hospital_name",               # new column name
    value="Thomas-jefferson-university-hospital"    # constant value for every row
)
df3
```

```python
df3.dropna(inplace=True)

# Remove duplicate rows
df3.drop_duplicates(inplace=True)
```

```python
df3
```

```python
null_counts = df3.isnull().sum()
print("Count of null values per column:")
print(null_counts)
```

```python
duplicate_count = df3.duplicated().sum()

print(f"Number of duplicate records: {duplicate_count}")
```

```python
df3.to_csv('Thomas-jefferson-university-hospital.csv', index=0)
```
